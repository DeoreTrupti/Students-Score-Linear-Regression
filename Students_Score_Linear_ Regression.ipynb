# Importing all libraries required in this notebook

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
import seaborn as sns
%matplotlib inline
```

# Reading data from remote link

```python
data_set = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print("Data imported successfully")
```

```python
data_set.head()
```

```python
data_set.isnull().sum()
```

# plotting the scatter diagram

```python
sns.scatterplot(y=data_set['Scores'],x=data_set['Hours'])
plt.title('Marks vs Study Hours',size=15)  
plt.xlabel('Marks Percentage',size=12)  
plt.ylabel('Marks Studied',size=12)  
plt.show()
```

# Plotting the regression line

```python
sns.regplot(y=data_set['Scores'],x=data_set['Hours'])
plt.title('Marks vs Study Hours',size=15)  
plt.xlabel('Marks Percentage',size=12)  
plt.ylabel('Marks Studied',size=12)  
plt.show()
print(data_set.corr())
```

# Preparing the data

```python
X = data_set.iloc[:, :-1].values  
y =data_set.iloc[:, 1].values 
```

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.25, random_state=0) 
```

#75% traning data and 25% test data

# Training the Algorithm

```python
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 

print("Training complete.")
```

# Making Predictions

```python
print(X_test) # Testing data - In Hours
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
# Comparing Actual vs Predicted
df = pd.DataFrame({'Hours':[i[0] for i in X_test],'Predicted Marks':[k for k in y_pred]})  
df 
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df 
```

```python
plt.scatter(x=X_test,y=y_test,color='red')
plt.plot(X_test,y_pred,color='black')
plt.title('Actual vs Predicted',size=15)  
plt.xlabel('Marks Percentage',size=12)  
plt.ylabel('Marks Studied',size=12)  
plt.show()
print(data_set.corr())
```

# Evaluating the model

```python
from sklearn import metrics as m
print('Mean Absolute Error:', m.mean_absolute_error(y_test, y_pred)) 
```
